### Univariate, Bivariate, Multivariate Analaysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('iris.csv')
data
```

```python
data.head()
```

```python
data.shape
```

##### Unvariate Analysis :
- We picked up only one feature, based on this feature we will try to determine what the output is.

```python
d0 = data.loc[data["target"]==0]
d0
```

```python
d1 = data.loc[data["target"]==1]
d1
```

```python
d2 = data.loc[data["target"]==2]
d2
```

```python
plt.plot(d0['sepal_length'],np.zeros_like(d0['sepal_length']),"o")
plt.plot(d1['sepal_length'],np.zeros_like(d1['sepal_length']),"o")
plt.plot(d2['sepal_length'],np.zeros_like(d2['sepal_length']),"o")
plt.xlabel("petal length")
plt.show()
```

##### Bivariate Analysis : 
- We picked up two feature, based on these feature we will try to determine what the output is.

```python
sns.FacetGrid(data,hue="target", size=5).map(plt.scatter,"petal_length","sepal_width").add_legend();
plt.show()
```

##### Multivariate Analysis : 
- We picked up more than two feature, based on these feature we will try to determine what the output is.

```python
sns.pairplot(data,hue="target",size=3)
```
