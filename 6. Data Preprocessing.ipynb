##### outlier detection and treating

```python
b = pd.read_csv("breast_cancer.csv")
b.drop(['Unnamed: 0'],axis=1, inplace=True)
b
```

```python
from scipy import stats
z_score = np.abs(stats.zscore(b))
z_score
```

```python
threshold = 3 
```

```python
out = np.where(z_score>threshold)
out
```

```python
outlier = pd.DataFrame(out).T
outlier.columns = ["Row","Column"]
outlier
```

```python
outlier.Column.value_counts()
```

```python
b.columns
```

Outliers are present in all the columns 0 to 29 except 27

```python
import seaborn as sns
```

```python
sns.boxplot(x = b['mean area'])
```

```python
Q1, Q3 = np.percentile(b["mean area"],[25,75])
print("Q1 = ",Q1)
print("Q3 = ",Q3)
IQR = Q3 - Q1
print("IQR = ",IQR)
low = Q1 - (1.5*IQR)
upr = Q3 + (1.5*IQR)
print("Lower Bound = ",low)
print("Upper Bound = ",upr)
mea = b["mean area"].mean()
print("Mean values of mean area = ",mea)
```

```python
b["mean area"] = np.where(b["mean area"] < low, mea, b["mean area"])
b["mean area"] = np.where(b["mean area"] > upr, mea, b["mean area"])
```

```python
print(np.where(b["mean area"] < low))
print(np.where(b["mean area"] > upr))
```

##### Handle Imbalanced data : 
- When difference b/w majority class and minority class is huge than dataset is imbalanced
- How to make dataset imbalanced
    - using Under Sampling : loss sample from majority class and averaging both class's value
    - using Over Sampling : add duplicate sample in minority class and averaging both class's value
    - using SMOTE : providing synthetic sample to minority class and averaging both class's value

```python
data_ = data.select_dtypes(include=['int64'])
data_
```
