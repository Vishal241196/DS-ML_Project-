```python
import numpy as np
import pandas as pd
```

```python
iris = pd.read_csv("iris.csv")
iris
```

```python
X = iris.iloc[:,:-1]
y = iris["target"]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
```

```python
from sklearn.model_selection import GridSearchCV
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 10, test_size = 0.3)
```

### 1. Using Logistic Regression

```python
log = LogisticRegression()
para_log = {'penalty' : ['l1','l2'],
            'C' : [1,2,3],
            'random_state': [2,5,10]}
grid_log = GridSearchCV(log,para_log,cv=2,verbose=True)
grid_log.fit(X_train,y_train)
```

```python
print(grid_log.best_estimator_)
print(grid_log.best_score_)
print(grid_log.best_params_)
```

```python
Log = LogisticRegression(penalty='l2',C = 3, random_state=2)
Log.fit(X_train,y_train)
print(Log.score(X_train,y_train))
Log_ypredict = Log.predict(X_test)
```

```python
accuracy_score(y_test,Log_ypredict)
```

### 2. Using KNN

```python
knn = KNeighborsClassifier()
para_knn = {'n_neighbors' : [5,10,15],
            'leaf_size' : [10,20,30],
            'p' : [1,2,3],
            'metric' : ['minkowski']}
```

```python
grid_knn = GridSearchCV(knn,para_knn,cv=2,verbose=True)
grid_knn.fit(X_train,y_train)
```

```python
print(grid_knn.best_params_)
print(grid_knn.best_score_)
print(grid_knn.best_estimator_)
```

```python
KNN = KNeighborsClassifier(leaf_size=10,metric='minkowski',n_neighbors=10, p=1)
KNN.fit(X_train,y_train)
print(KNN.score(X_train,y_train))
KNN_ypredict = KNN.predict(X_test)
```

```python
accuracy_score(y_test,KNN_ypredict)
```

### 3. Using Decision Tree

```python
dt = DecisionTreeClassifier()
para_dt = {'criterion' : ['gini','entropy'],
           'max_depth' : range(1,10),
           'min_samples_split' : range(1,10),
           'min_samples_leaf' : range(1,5)}
grid_dt = GridSearchCV(dt, para_dt, cv = 2, verbose=True)
grid_dt.fit(X_train,y_train)
```

```python
print(grid_dt.best_score_)
print(grid_dt.best_params_)
print(grid_dt.best_estimator_)
```

```python
DT = DecisionTreeClassifier(criterion='entropy',
                            max_depth=7,
                            min_samples_leaf=1,
                            min_impurity_split=2,
                            random_state=10)
DT.fit(X_train,y_train)
print(DT.score(X_train,y_train))
DT_ypredict = DT.predict(X_test)
```

```python
accuracy_score(y_test,DT_ypredict)
```

### 4. Using Random Forest

```python
rf = RandomForestClassifier()
para_rf = {'n_estimators' : [100,200],
           'min_samples_leaf' : [2,3,5],
           'min_samples_split' : [2,3,5],
           'max_depth' : [10,30,50]}
grid_rf = GridSearchCV(rf, para_rf, cv = 2, verbose = True)
grid_rf.fit(X_train,y_train)
```

```python
print(grid_rf.best_score_)
print(grid_rf.best_params_)
print(grid_rf.best_estimator_)
```

```python
RF = RandomForestClassifier(n_estimators=100,
                           max_depth=10,
                           min_samples_leaf=2,
                           min_samples_split=2)
RF.fit(X_train,y_train)
print(RF.score(X_train,y_train))
RF_ypredict = RF.predict(X_test)
```

```python
accuracy_score(y_test,RF_ypredict)
```

### 5. Using SVM

```python
svm = SVC()
para_svm = {'kernel':['rbf','linear'],
           'C':[1,50,100,500],
           'gamma':[0.0001,0.01,1.0],
           'random_state':[5,10]}
grid_svm = GridSearchCV(svm, para_svm, cv = 2, verbose = True)
grid_svm.fit(X_train,y_train)
```

```python
print(grid_svm.best_score_)
print(grid_svm.best_params_)
print(grid_svm.best_estimator_)
```

```python
SVC = SVC(C=1, gamma=0.0001, kernel='linear', random_state=5)
SVC.fit(X_train,y_train)
print(SVC.score(X_train,y_train))
SVC_ypredict = SVC.predict(X_test)
```

```python
accuracy_score(y_test,SVC_ypredict)
```

### 6. Using XGBoosT

```python
xgb = XGBClassifier()
para_xgb = {'nthread':[2,4,6], 
            'learning_rate': [.03, 0.05, .07], 
            'max_depth': [5, 6, 7],
            'n_estimators': [100,200]}
grid_xgb = GridSearchCV(xgb, para_xgb, cv = 2, verbose = True)
grid_xgb.fit(X_train,y_train)
```

```python
print(grid_xgb.best_score_)
print(grid_xgb.best_params_)
print(grid_xgb.best_estimator_)
```

```python
XGB = XGBClassifier(nthread=2,
                    learning_rate=0.03,
                    max_depth=5,
                    n_estimators=200)
XGB.fit(X_train,y_train)
print(XGB.score(X_train,y_train))
XGB_ypredict = XGB.predict(X_test)
```

```python
accuracy_score(y_test,XGB_ypredict)
```
