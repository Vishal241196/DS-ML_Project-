## sort series

```python
import pandas as pd
```

```python
dic = {
    1:'one',
    3:'three',
    2:'two',
    4:'four'
}
```

```python
seri = pd.Series(dic , name='numbers')
```

```python
seri
```

```python
#we do sorting from 2 ways, by index or by values
seri.sort_index()#by index:-sort according to index,it doesn't sort actual series
```

```python
seri.sort_index(ascending=True)
```

```python
seri.sort_index(ascending=False)
```

```python
seri#actual series is not sort
```

```python
seri.sort_index(ascending=True, inplace=True)#for sorting actual series
```

```python
seri
```

```python
seri.sort_values()#by values:-sort according to values(alphabetically)
```

```python
seri.sort_values(ascending=True)
```

```python
seri.sort_values(ascending=False)
```

```python
seri.sort_values(ascending=True, inplace=True)
```

```python
seri
```

## Largest and smallest 

```python
import pandas as pd
```

```python
guitar_dic = {
             'Yamsha F310'          : 8500,
             'Faith Apollo Venus'   : 30000,
             'Taylor 214 CE'        : 250000,
             'Baton Rounge X6C'     : 40000,
             'PRS SE'               : 42000,
             'Lakewood'             : 300000
            }
```

```python
guitar_price = pd.Series(guitar_dic, name= 'price')
```

```python
guitar_price
```

```python
guitar_price.nsmallest()#gives data of top5 smallest price guitar
```

```python
guitar_price.nsmallest(n=2)
```

```python
guitar_price.nsmallest(n=6)
```

```python
guitar_price.nsmallest().index[0]#gives name of smallest price guitar
```

```python
guitar_price.nlargest()#top5 largest price guitar
```

```python
guitar_price.nlargest(n=3)
```

```python
guitar_price.nlargest().index[0]
```

```python
guitar_price.idxmin()#gives index of smallest value
```

```python
guitar_price.idxmax()#gives index of largest value
```
